```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid", font_scale=1.1)
df = pd.read_csv(
    "emissions.csv",
    sep=";",               
    decimal=",",           
    encoding="utf-8-sig"   
)


print(df.shape)
df.head()
```

```python
corr = df.corr(method="pearson", numeric_only=True)
print(corr.shape)
corr.head()
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(
    corr,
    annot=True,         # показывать значения
    fmt=".2f",
    cmap="RdYlGn",      # цветовая схема
    center=0,           # нулевая корреляция = белый цвет
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)
plt.title("Корреляционная матрица параметров эмиссий", fontsize=16, pad=15)
plt.tight_layout()

plt.savefig("figures/correlation_heatmap.png", dpi=300, bbox_inches="tight")

plt.show()

corr.to_csv("data/correlation_matrix.csv", index=True)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# пары для визуализации (добавь любые другие)
pairs = [
    ('масс.конц-ция SO2 Влажная', 'масс.конц-ция CO Влажная'),
    ('масс.конц-ция O2 Влажная',  'масс.конц-ция CO Влажная'),
    ('масс.конц-ция NO Влажная',  'масс.конц-ция CO Влажная'),
    ('масс.конц-ция SO2 Влажная', 'масс.конц-ция NO Влажная'),
]

for x_col, y_col in pairs:
    plt.figure(figsize=(6,5))
    
    # регрессионная линия + точки
    sns.regplot(
        data=df,
        x=x_col, y=y_col,
        scatter_kws={'alpha':0.5, 's':30, 'color':'#007ACC'},
        line_kws={'color':'#FF6B35', 'linewidth':2}
    )
    
    # корреляция Пирсона
    r, p = pearsonr(df[x_col], df[y_col])
    
    plt.title(f"{y_col} vs {x_col}\nПирсон r = {r:.2f}, p = {p:.1e}",
              fontsize=13, weight='bold')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()


```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(
    data=df,
    x='масс.конц-ция SO2 Влажная',
    y='масс.конц-ция CO Влажная',
    alpha=0.6, s=30, color='#007ACC', edgecolor='none'
)
sns.regplot(
    data=df,
    x='масс.конц-ция SO2 Влажная',
    y='масс.конц-ция CO Влажная',
    scatter=False, color='#FF6B35', line_kws={'linewidth':2}
)
plt.title("CO vs SO₂ — линейная зависимость", fontsize=14, weight='bold')
plt.xlabel("SO₂, влажная концентрация")
plt.ylabel("CO, влажная концентрация")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(
    data=df,
    x='масс.конц-ция O2 Влажная',
    y='масс.конц-ция CO Влажная',
    alpha=0.5, s=25, color='#222222'
)
sns.regplot(
    data=df,
    x='масс.конц-ция O2 Влажная',
    y='масс.конц-ция CO Влажная',
    scatter=False, color='#00A878', line_kws={'linewidth':2}
)
plt.title("CO vs O₂ — обратная зависимость", fontsize=14, weight='bold')
plt.xlabel("O₂, влажная концентрация")
plt.ylabel("CO, влажная концентрация")
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')

p = ax.scatter(
    df['масс.конц-ция SO2 Влажная'],
    df['масс.конц-ция O2 Влажная'],
    df['масс.конц-ция CO Влажная'],
    c=df['масс.конц-ция CO Влажная'],
    cmap='plasma', alpha=0.7, s=20
)
ax.set_xlabel("SO₂, влажная")
ax.set_ylabel("O₂, влажная")
ax.set_zlabel("CO, влажная")
ax.set_title("3D зависимость CO от SO₂ и O₂", fontsize=14, weight='bold')
fig.colorbar(p, ax=ax, shrink=0.5, aspect=10, label='CO уровень')
plt.tight_layout()
plt.show()
```
